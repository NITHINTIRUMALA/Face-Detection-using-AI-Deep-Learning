### Write a program for marks grading
#### >70 marks 'A+' Grade  
#### >60 marks 'B' Grade 
#### >50 marks 'C' Grade 
#### >40 marks 'D' Grade 
#### <40 marks Fail

#### Take Student name, marks as input and output as below
#### "Student_Name passed with "Grade"

```python

name = input('Student name : ')
marks = int(input ('marks : '))

if marks > 70:
    print(name, 'passed with A+')
elif marks > 60:
    print(name, 'passed with B')
elif marks > 50:
    print(name, 'passed with C')
elif marks > 40:
    print(name, 'passed with D')
else:
    print(name, 'failed')
```

### Write a program to print Geometric progression take the inital value(a) and common ratio(r)

```python
# a, a*r^1, a*r^2 .....
```

```python
a = int(input('a : '))
r = int(input('r : '))

for i in range(20):
    print(a*(r**i), end=', ')
```

### Write a program to print Fibonacci series till first 50 terms

```python
 1, 1 , 2, 3, 5, ...
```

```python
a = 0
b = 1
print(a, b, sep=', ', end=', ')
for i in range(50):
    a, b = b, a+b
    print(b, sep=', ', end=', ')
```
