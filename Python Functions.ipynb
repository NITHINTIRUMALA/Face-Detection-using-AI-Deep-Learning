```python
#Function with passing arguments and returning values

def function(a : float, b : int):
    prod = a*b
    flr_div = a//b
    return{b,a, prod, flr_div}

print('Enter x value')
x = int(input())
print('Enter y value')
y = int(input())
res = function(x,y)
print(res, type(res))
print(res)

a,b,c,d = res
print(res,'\n',a,d,c,b)
```

## Write a program for creating a calculator

### Take operands and operator as inputs and process the resuts

```python
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a/b

a = int(input('Enter value a : '))
b = int(input('Enter value b : '))

print('Sum : ', add(a,b))
print('diff : ', sub(a,b))
print('product : ', mul(a,b))
print('division : ', div(a,b))
```
