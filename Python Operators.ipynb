# Assignment operator

### =, +=, -=, *=, /=, **=, //=, %=

```python
print('Enter x value"')
x = int(input())
z = x
print('z = x : ','z=>',z)
z += x # z= z+x
print('z += x : ','z=>',z)
z -= x
print('z -= x : ','z=>',z)
z *= x
print('z *= x : ','z=>',z)
z /= x
print('z /= x : ','z=>',z)
z **= x
print('z **= x : ','z=>',z)
z //= x
print('z //= x : ','z=>',z)
z %= x
print('z %= x : ','z=>',z)
```
